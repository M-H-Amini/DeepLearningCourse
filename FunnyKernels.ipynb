In this project, I am going to see some funny kernels in image processing. The following kernels are implemented:


*   Blur
*   Sharpen
*   Outline
*   Emboss
*   Right Sobel
*   Top Sobel



First of all, let's import module...

```
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

Now time for importing the image, resizing and showing it...

```
image = cv2.resize(cv2.imread('photo_2019-05-11_01-57-22.jpg', 0), (400, 300))
plt.imshow(image, cmap='gray');
```

# Blur Kernel


```
blur = np.array([[0.0625, 0.125, 0.0625], [0.125, 0.25, 0.125], [0.0625, 0.125, 0.0625]])
blurred_image = cv2.filter2D(image, -1, blur)
plt.figure()
plt.imshow(blurred_image, cmap='gray');
```

# Sharpen Kernel

```
sharpen = np.array([[0, -1, 0], [-1, 5, -1], [0, -1, 0]])
sharpened_image = cv2.filter2D(image, -1, sharpen)
plt.figure()
plt.imshow(sharpened_image, cmap='gray');
```

# Outline Kernel

```
outline = np.array([[-1, -1, -1], [-1, 8, -1], [-1, -1, -1]])
outlined_image = cv2.filter2D(image, -1, outline)
plt.figure()
plt.imshow(outlined_image, cmap='gray');
```

# Emboss Kernel

```
emboss = np.array([[-2, -1, 0], [-1, 1, 1], [0, 1, 2]])
embossed_image = cv2.filter2D(image, -1, emboss)
plt.figure()
plt.imshow(embossed_image, cmap='gray');
```

# Right Sobel Kernel

```
right_sobel = np.array([[1, 0, -1], [2, 0, -2], [1, 0, -1]])
right_sobeled_image = cv2.filter2D(image, -1, right_sobel)
plt.figure()
plt.imshow(right_sobeled_image, cmap='gray');
```

# Top Sobel Kernel

```
top_sobel = np.array([[1, 2, 1], [0, 0, 0], [-1, -2, -1]])
top_sobeled_image = cv2.filter2D(image, -1, top_sobel)
plt.figure()
plt.imshow(top_sobeled_image, cmap='gray');
```
